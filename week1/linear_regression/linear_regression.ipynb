```python
# import the libraries
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# load the data
df = pd.read_csv('home_dataset.csv')
df.head(5)
```

```python
# Extract the independent and dependent variables
#the data from the column named “HouseSize” is being added to the house_size variable.
house_size = df['HouseSize'].values
house_price = df['HousePrice'].values
```

```python
import statistics
# Calculate the mean of the house size and house price
print("Mean of House Size: ", statistics.mean(house_size))
print("Mean of House Size: ", statistics.mean(house_price))
```

```python
# visualize the data
plt.scatter(house_size, house_price, marker='o', color='red')
plt.title('House Price vs House Size')
plt.xlabel ('House Size(sqft)')
plt.ylabel('House Price($)')
plt.show()
```

## Training Test

A [train-test](https://en.wikipedia.org/wiki/Training,_validation,_and_test_data_sets) splits the generated data into training and testing sets using the `train_test_split()` function from Sklearn.

```python
# Split the data into training and testing data
# 80% of the data will be used for training and 20% will be used for testing
x_train, x_test, y_train, y_test = train_test_split(house_size, house_price, test_size=0.2, random_state=42)
```

```python
# Reshape the data to fit the model
x_train = x_train.reshape(-1, 1)
x_test = x_test.reshape(-1, 1)
```

```python
#create and train the model
model = LinearRegression()
model.fit(x_train, y_train)
```

## Prediction 

```python
# Predict the house price for the test data
y_pred = model.predict(x_test)

# Visualize the predicted house price
plt.scatter(x_test, y_test, marker='o', color='red', label='Actual Price')
plt.plot(x_test, y_pred, color='blue', linewidth=2, label='Predicted Price')
plt.title('House Price vs House Size (Test Data)')
plt.xlabel('House Size(sqft)')
plt.ylabel('House Price($)')
plt.legend()
plt.show()

```
